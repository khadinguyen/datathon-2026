## Data Preprocessing

The first thing that we did to the data was load the files and combine them into a single DataFrame\. We then preprocessed the tweet text by removing retweet indicators, links, mentions, hashtags, and any extra punctuation\. We also converted everything to lowercase and removed any unnecessary extra characters\. We kept track of various features for our analysis such as character and word count, mentions, hashtags, and the amount of exclamation and question marks\. Finally, all the processed data and summary statistics were saved for exploratory analysis\.

## IDF, TF, and Vectorization

Individual characters, common words like the and a, and even items like emojis, such as laugh emoji appear a lot in tweets with little to no sentimental value\. To compensate for the inherently differing baselines for these different kinds of terms, we can use Inverse Document Frequency is normalised with a factor which measures how common a term is throughout the whole collection of tweets\. The IDF has a high value for terms which occur in very few of the tweets and a low value for terms and characters which occur in almost all of the tweets\. For the tweets we have looked at, this normalisation is particularly effective for the emojis\. While there are many instances of laugh emoji, the sentiment providing emojis like flushed and relaxed are far rarer\. These words appear less frequently in the dataset and therefore have a higher IDF weight, meaning that when they do appear, they will carry a lot of weight in the sentiment prediction\.

We also used Term Frequency and Vectorization to analyze tweets\. By turning the words and emojis in a tweet into vectors using vectorization, we can use the bag of words process to count the unique words and ignore common words\. Once every tweets is converted to a vector of word counts and other factors, a machine learning model can compare data and find patterns like heart emoji correlating to the word Starbucks\.

## Model Training

### ML General Model

Logistic Regression Model : We found that emoji prediction from tweet text is a complex task due to the subtle and overlapping nature of emotional expressions in social media\. To establish a baseline, we implemented a Logistic Regression model using TF\-IDF vectorized text as the primary feature representation\. Logistic Regression was chosen for its simplicity, interpretability, and effectiveness in high\-dimensional sparse data\. We also experimented with class weighting to address imbalance across emoji classes\. Despite these efforts, the model achieved moderate performance, with accuracy significantly below our target threshold\. This suggests that linear decision boundaries are insufficient for capturing the complex relationships between textual features and emoji labels\. Additionally, the similarity in language used across different emojis further limited the model’s ability to distinguish between classes\. Overall, while Logistic Regression provided a solid baseline, it lacked the capacity to model deeper semantic relationships required for this task\.

Naive Bayes Model :To further explore lightweight text classification approaches, we implemented a Multinomial Naive Bayes model using TF\-IDF features\. Naive Bayes assumes feature independence and is particularly efficient for text\-based tasks, making it a strong candidate for baseline comparison\. However, due to the independence assumption, the model struggles to capture interactions between words and context within tweets\. As a result, performance was relatively low compared to other models, with poor differentiation across similar emoji classes\. The model tended to favor more frequent classes, highlighting the impact of class imbalance in the dataset\. Additionally, the noisy and informal nature of Twitter text—such as slang, abbreviations, and sarcasm—further reduced the effectiveness of this probabilistic approach\. While Naive Bayes provided fast and interpretable results, it ultimately lacked the complexity needed for accurate emoji prediction\.

Building on our baseline models, we implemented a Linear Support Vector Machine \(SVM\) to better handle high\-dimensional text data\. SVM is well\-suited for classification tasks with sparse features, as it seeks to maximize the margin between classes\. Using TF\-IDF features and experimenting with class weights, we observed improved performance compared to Logistic Regression and Naive Bayes\. The model demonstrated stronger ability to separate certain emoji classes, particularly those with more distinct language patterns\. However, overall performance remained limited due to the inherent overlap in emoji meanings and the noisy nature of tweet data\. Despite its robustness, the linear SVM still relies on surface\-level features and cannot fully capture deeper contextual relationships in language\. This highlights the limitations of traditional machine learning approaches for complex natural language understanding tasks\.

### XG Model

We observed that predicting emojis directly from tweet text is inherently challenging due to the high variability and contextual nature of language on social media\. Emojis often represent nuanced emotional states that can overlap significantly in meaning, making it difficult for traditional machine learning models to distinguish between them\. To address this, we implemented an XGBoost model, leveraging its ability to capture non\-linear relationships and interactions between features such as TF\-IDF representations, sentiment scores, and additional engineered features like word counts and craving indicators\. Through hyperparameter tuning, we optimized key parameters including the number of estimators, regularization strength, and sampling ratios to improve generalization\. Despite these efforts, the XGBoost model achieved a test accuracy of approximately 51%, indicating only moderate predictive performance\. This suggests that while XGBoost is powerful for structured data, it struggles with the high dimensionality and sparsity of textual features in this task\. Additionally, the overlapping semantic meaning of emojis and the noisy nature of Twitter data further limited the model’s ability to learn clear decision boundaries\. Overall, the results highlight that more advanced natural language understanding techniques, such as deep learning or transformer\-based models, may be necessary to significantly improve performance in emoji prediction tasks\.

### SVM Model

From the previous models, we discovered that the accuracies and macro f1 scores were very low with values barely reaching 50%\. Intuitively, the nature of emoji usage and social media posts are inherently personal to each person where individuals use emojis in context\-dependent ways\.  As a result, simply predicting the exact emoji of a tweet can be challenging as multiple emojis can convey the similar emotions\. For example, heart\-eyes, winking, and blushing emojis can all be used to express affection, making it difficult for the model to distinguish between them\. To alleviate this, for our SVM model we altered the emoji labeling to map onto a sentiment where emojis that express positive or negative emotions will be labeled identically, respectively\. We then used this sentiment label to train our predictor\.

Our first SVM model accounts for all the transformed features in the dataset\. Our second SVM model only used the TF\-IDF cleaned text as features in the model\. The second model aims to test if just the vectorized text is sufficient enough to predict sentiment when compared to the model with all the features\.  Using grid search techniques to tune the parameters \(regularization and class weight\), we found these parameters to be the best performing\.

Overall, by using sentiment labeling instead of singular emoji labeling, the model is better able to predict the emotions of the tweet\.
