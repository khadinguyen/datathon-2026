# Report Analysis

## Introduction: Dataset

For our project we decided to take on both the StrataScratch and Omni challenge\. With the StrataScratch challenge, we were given a dataset of tweets that used an emoji \(with the emoji omitted\), and the emojis that correspond to each tweet\. Using a model that can predict emoji's based on text, it can determine the emotions behind each post\. Business can then leverage this data to better understand customer response\. For food and retail brands, the model can provide real\-time insights into customer emotions and cravings which can then help them tailor marketing campaigns\. Social media platforms may also benefit from this model as it can automatically tag emotional tone in tweets to improve content algorithms and recommendations\.

## Data Preprocessing

The first thing that we did to the data was load the files and combine them into a single DataFrame\. We then preprocessed the tweet text by removing retweet indicators, links, mentions, hashtags, and any extra punctuation\. We also converted everything to lowercase and removed any unnecessary extra characters\. We kept track of various features for our analysis such as character and word count, mentions, hashtags, and the amount of exclamation and question marks\. Finally, all the processed data and summary statistics were saved for exploratory analysis\.

## Data Exploration and Visualizations

To analyze the dataset, we created visualizations of various statistics\. We first looked at how often a brand name was mentioned, and found that Starbucks was mentioned the most \(over 60,000 times\), with other brands mentioned less\.

<img src="image-20260405-031447.png" width="" align="" />

However, there was no strong correlation between popularity and positivity\. For example, Chipotle had the highest average text sentiment while being mentioned less times while Walmart was mentioned more but had the lowest text sentiment\.

<img src="image-20260405-031655.png" width="" align="" />

Comparing brand\-level sentiment with emoji sentiment, it is shown that emoji sentiment is much more unpredictable as Five Guys and Wingstop have notably negative emoji sentiment while the text sentiment stays relatively neutral\. This suggests that the user's emotional expression through emojis contrasts the tone of their written text\.

<img src="image-20260405-031900.png" width="" align="" />

Looking at the emojis, the sob emoji was used the most \(50,000 times\) which is then followed by the heart eyes and weary emojis\. This shows that food and retail tweets tend to skew toward strong emotional reactions\. 

<img src="outputs/figures/01_class_distribution.png" width="" align="" />

Looking at the top words for each emoji, it is dominated mostly by "bet" and "rt\." However, this is not surprising as "bet" was the most frequently used word in the entire dataset, with "rt" in second\. This means that their dominance in emoji distribution only represents their presence in the dataset as a whole and doesn't relate to any emojis\. More meaningful words are the secondary ones\. For example, the words "hungry" and "love" appear under the heart eyes emoji distribution and and "delicious" appears under the yum emoji distribution\. Despite the high presence of some words, the emoji word distribution can still carry some weight which is useful in prediction\.
